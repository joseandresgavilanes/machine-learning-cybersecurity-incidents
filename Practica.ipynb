```python
#Una columna interesante fue la del device infromation, al ser el user agent podiamos extraer tanto el browser como el dispositivo/os desde el que se envio.

import re
df['Device'] = None
df['Browser'] = None

patterns = [
    r'Windows',
    r'Linux',
    r'Android',
    r'iPad',
    r'iPod',
    r'iPhone',
    r'Mac',
]

def extract_device_or_os(user_agent):
    for pattern in patterns:
        match = re.search(pattern, user_agent, re.I)
        if match:
            return match.group()
    return 'Unknown'

# Función para extraer sistema operativo y navegador
def extraer_info_dispositivo(user_agent):
    so_match = re.search(r'\(([^)]+)\)', user_agent)
    navegador_match = re.search(r'^[^/]+', user_agent)
    
    sistema_operativo = so_match.group(1) if so_match else "Unknown"
    navegador = navegador_match.group(0) if navegador_match else "Unknown"
    
    sistema_operativo_identificado = extract_device_or_os(sistema_operativo)
    
    return sistema_operativo_identificado, navegador

df[['Device', 'Browser']] = df['Device Information'].apply(
    lambda x: pd.Series(extraer_info_dispositivo(x))
)

df.drop(['Device Information'], axis=1, inplace=True)
df.head()
```

```python
#La columna de Geo-location Data podemos dividirla en ciudad y estado.

df[['City', 'State']] = df['Geo-location Data'].str.split(', ', expand=True)
df.drop(['Geo-location Data'], axis=1, inplace=True)
```

```python
# Analizando el resto de columnas nos dimos cuenta de que las siguientes no iban a ser de gran relevancia
df.drop(['Timestamp', 'Payload Data', 'Source Port', 'Destination Port', 'IDS/IPS Alerts', 'Source IP Address','Destination IP Address','User Information', 'Firewall Logs', 'Proxy Information','Log Source'], axis=1, inplace=True)
```

```python
# Rellenamos los nulos encontrados en Malware indicators con el valor de 'None Detected'
df['Malware Indicators'].fillna('None Detected', inplace=True)
# Rellenamos los nulos encontrados en Malware indicators con el valor de 'No Alert'
df['Alerts/Warnings'].fillna('No Alert/Warning', inplace=True)

#verificamos
print(df["Malware Indicators"].unique())
print(df["Alerts/Warnings"].unique())
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel


# Selección de características usando Random Forest
selector = SelectFromModel(RandomForestClassifier(n_estimators=100, random_state=42))
selector.fit(X_train, y_train['action_taken'])
selected_features = X_train.columns[(selector.get_support())]
X_train = X_train[selected_features]
X_test = X_test[selected_features]

print("Características seleccionadas Action Taken:", selected_features)

```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report

models = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier()
}

# Entrenamiento y evaluación de cada modelo
for model_name, model in models.items():
    model.fit(X_train, y_train['action_taken'])
    scores = cross_val_score(model, X_train, y_train['action_taken'], cv=5)
    print(f"{model_name} Cross-Validation Score: {scores.mean()}")

    # Predicción y reporte
    y_pred = model.predict(X_test)
    print(f"{model_name} Classification Report:")
    print(classification_report(y_test['action_taken'], y_pred))
```
